```python
import pandas as pd
```

```python
df0 = pd.read_csv('Crashes_in_DC_0.csv')
df1 = pd.read_csv('Crashes_in_DC_1.csv')
df2 = pd.read_csv('Crashes_in_DC_2.csv')
df3 = pd.read_csv('Crashes_in_DC_3.csv')
df4 = pd.read_csv('Crashes_in_DC_4.csv')
df5 = pd.read_csv('Crashes_in_DC_5.csv')
```

```python
rawCrash = pd.concat([df0, df1, df2, df3, df4, df5])
```

```python
rawCrash.columns
```

```python
rawCrash['REPORTDATE'] = pd.to_datetime(rawCrash['REPORTDATE']).dt.tz_convert(None)
```

```python
rawCrash['DATE'] = pd.to_datetime(rawCrash['REPORTDATE']).dt.date
rawCrash['TIME'] = pd.to_datetime(rawCrash['REPORTDATE']).dt.time
```

```python
rawCrash['DATE'] = pd.to_datetime(rawCrash['DATE'])
```

```python
rawCrash = rawCrash.loc[(rawCrash['DATE'].dt.year > 2014) & (rawCrash['DATE'].dt.year < 2020)]
```
